```python
import csv

def main():
    f=open('202303_Seoul_Subway.csv','r', encoding = 'utf-8')
    data = csv.reader(f,delimiter=',')
    header=next(data)
    max_1=''
    min_1=''
    max_2=''
    min_2=''
    max_3=''
    min_3=''
    max_4=''
    min_4=''
    max_num_1=0
    min_num_1=99999999
    max_num_2=0
    min_num_2=99999999
    max_num_3=0
    min_num_3=99999999
    max_num_4=0
    min_num_4=99999999
    
    for row in data:
        str = row[1]
        row[-3] = int(row[-3])
        row[-2] = int(row[-2])
        if str[0]=='1':
            if row[-2]+row[-3]>max_num_1:
                max_num_1 = row[-2]+row[-3]
                max_1=row[3]
            if row[-2]+row[-3]<min_num_1:
                min_num_1 = row[-2]+row[-3]
                min_1=row[3]
                
        if str[0]=='2':
            if row[-2]+row[-3]>max_num_2:
                max_num_2 = row[-2]+row[-3]
                max_2=row[3]
            if row[-2]+row[-3]<min_num_2:
                min_num_2 = row[-2]+row[-3]
                min_2=row[3]
        
        if str[0]=='3':
            if row[-2]+row[-3]>max_num_3:
                max_num_3 = row[-2]+row[-3]
                max_3=row[3]
            if row[-2]+row[-3]<min_num_3:
                min_num_3 = row[-2]+row[-3]
                min_3=row[3]
                
        if str[0]=='4':
            if row[-2]+row[-3]>max_num_4:
                max_num_4 = row[-2]+row[-3]
                max_4=row[3]
            if row[-2]+row[-3]<min_num_4:
                min_num_4 = row[-2]+row[-3]
                min_4=row[3]
                
    print("*** Subway Report for Seoul on March 2023 ***")
    print("Line 1:")
    print("Busiest Station: {0:s} ({1:d})".format(max_1, max_num_1))
    print("Least used Station: {0:s} ({1:d})".format(min_1, min_num_1))
    print("Line 2:")
    print("Busiest Station: {0:s} ({1:d})".format(max_2, max_num_2))
    print("Least used Station: {0:s} ({1:d})".format(min_2, min_num_2))
    print("Line 3:")
    print("Busiest Station: {0:s} ({1:d})".format(max_3, max_num_3))
    print("Least used Station: {0:s} ({1:d})".format(min_3, min_num_3))
    print("Line 4:")
    print("Busiest Station: {0:s} ({1:d})".format(max_4, max_num_4))
    print("Least used Station: {0:s} ({1:d})".format(min_4, min_num_4))
    
if __name__=="__main__":
    main()
```
